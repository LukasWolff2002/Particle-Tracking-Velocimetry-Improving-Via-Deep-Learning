```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import numpy as np  # Para el cálculo del ángulo

# Cargar el modelo y los pesos entrenados
model = YOLO("C:/Users/MBX/Desktop/PTV/YOLO/runs/segment/train/weights/best.pt")

# Aplicar el modelo sobre la imagen
results = model.predict(
    source='C:/Users/MBX/Desktop/PTV/YOLO/Sopa-de-Tallarines-2/test/images/Basler_acA1440-220uc__40343433__20241204_143241295_0009_bmp.rf.a422c33bc4c70bfdee54c21faaffe829.jpg',
    save=True,
    conf=0.25
)

# Extraer los datos necesarios y convertir a la CPU
boxes = results[0].boxes.xyxy.cpu().numpy()  # Coordenadas de las cajas (x1, y1, x2, y2)
scores = results[0].boxes.conf.cpu().numpy()  # Confianza de las detecciones
classes = results[0].boxes.cls.cpu().numpy()  # Clases de las detecciones

# Calcular los centroides, largo máximo y ángulo en el plano XY
centroids = []
max_lengths = []
angles = []
for box in boxes:
    x1, y1, x2, y2 = box
    cx = (x1 + x2) / 2  # Coordenada X del centroide
    cy = (y1 + y2) / 2  # Coordenada Y del centroide
    width = x2 - x1     # Ancho de la caja
    height = y2 - y1    # Altura de la caja
    max_length = max(width, height)  # Largo máximo (mayor dimensión)
    angle = np.degrees(np.arctan2(cy, cx))  # Ángulo en grados con respecto al eje X

    centroids.append((cx, cy))
    max_lengths.append(max_length)
    angles.append(angle)

# Crear un DataFrame con los resultados
data = {
    "Confianza": scores,
    "Centroide (cx, cy)": centroids,
    "Largo Máximo": max_lengths,
    "Ángulo (grados)": angles,
}
df = pd.DataFrame(data)

# Mostrar los resultados
print(df)

# Visualizar las imágenes lado a lado
# Rutas de las imágenes
original_image_path = 'C:/Users/MBX/Desktop/PTV/Cam0/Basler_acA1440-220uc__40343433__20241204_143241295_0000.bmp'
processed_image_path = 'runs/segment/predict2/Basler_acA1440-220uc__40343433__20241204_143241295_0009_bmp.rf.a422c33bc4c70bfdee54c21faaffe829.jpg'

# Leer las imágenes con OpenCV
original_image = cv2.imread(original_image_path)
processed_image = cv2.imread(processed_image_path)

# Convertir colores de BGR a RGB (para visualización con Matplotlib)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
processed_image = cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB)

# Crear la figura para mostrar las imágenes
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Imagen Original')
plt.axis('off')

# Mostrar la imagen procesada
plt.subplot(1, 2, 2)
plt.imshow(processed_image)
plt.title('Imagen Procesada')
plt.axis('off')

# Mostrar las imágenes lado a lado
plt.tight_layout()
plt.show()
```
